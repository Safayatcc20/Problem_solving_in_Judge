```python
for i in students:
    i['total'] = i["math"]  + i ["eng"] + i["science"]

for i in students:
    print(i['name'],'-> total',i['total'])
```

```python
highest = -1
topper_name = ""
for i in students:
    if i['total'] > highest:
        highest = i['total']
        topper_name = i['name']

print("Topper name: ", topper_name, 'with ', highest, 'marks')
```
